```python
from sentence_transformers import SentenceTransformer
from langchain.embeddings import HuggingFaceEmbeddings
embedded_model = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
data = dataset.to_pandas()
```

```python
from tqdm import tqdm

batch_size = 100
for i in tqdm(range(0, len(data), batch_size)):
    batch = data.iloc[i:i+batch_size]

    # Create IDs
    ids = [f"{row['doi']}-{row['chunk-id']}" for _, row in batch.iterrows()]
    texts = [row['chunk'] for _, row in batch.iterrows()]
    embeds = embedded_model.embed_documents(texts)

    metadata = [
        {'text': row['chunk'], 'source': row['source']} for _, row in batch.iterrows()
    ]

    # Prepare data for upsert
    vectors = list(zip(ids, embeds, metadata))
    index.upsert(vectors)
```

```python
index.describe_index_stats()
```

```python
from langchain_pinecone import PineconeVectorStore

text_field = "text"

vectorstore = PineconeVectorStore(
    index=index,
    embedding=embedded_model,
    text_key=text_field
)
```

```python
retriever = vectorstore.as_retriever(
    search_kwargs={"k": 3}  # Retrieve top 3 most relevant chunks
)
```

```python
from langchain.chains import ConversationalRetrievalChain

qa_chain = ConversationalRetrievalChain.from_llm(
    llm=chat,
    retriever=retriever,
    memory=memory,
    verbose=True  # Set to True to see the internal workings
)
```

```python
result = qa_chain({"question": "What is so special about Deepseek R1?"})
print(result["answer"])
```

```python
result = qa_chain.invoke({"question": "What are its reasoning capabilities?"})
print(result["answer"])
```
